```python
# Import the `random` module to shuffle the questions
import random

# Define the quiz data
quiz_data = [
    {"question": "What is the capital of France?", "answers": ["Paris", "London", "Berlin", "Rome"], "correct": 0},
    {"question": "Who painted the Mona Lisa?", "answers": ["Leonardo da Vinci", "Michelangelo", "Raphael", "Caravaggio"], "correct": 0},
    {"question": "What is the largest planet in our solar system?", "answers": ["Jupiter", "Saturn", "Uranus", "Neptune"], "correct": 0},
    {"question": "Who wrote Romeo and Juliet?", "answers": ["William Shakespeare", "Jane Austen", "Charles Dickens", "J.K. Rowling"], "correct": 0},
    {"question": "What is the chemical symbol for gold?", "answers": ["Au", "Ag", "Hg", "Pb"], "correct": 0},
    {"question": "Who was the first President of the United States?", "answers": ["George Washington", "Thomas Jefferson", "Abraham Lincoln", "Franklin D. Roosevelt"], "correct": 0},
    {"question": "What is the largest mammal on Earth?", "answers": ["Blue whale", "Fin whale", "Humpback whale", "Sperm whale"], "correct": 0},
    {"question": "Who painted the Starry Night?", "answers": ["Vincent van Gogh", "Paul Cézanne", "Claude Monet", "Pablo Picasso"], "correct": 0},
    {"question": "What is the smallest country in the world?", "answers": ["Vatican City", "Monaco", "Nauru", "Tuvalu"], "correct": 0},
    {"question": "Who wrote the famous novel 'To Kill a Mockingbird'?", "answers": ["Harper Lee", "F. Scott Fitzgerald", "Jane Austen", "J.K. Rowling"], "correct": 0}
]

# Shuffle the quiz data to randomize the order of questions
random.shuffle(quiz_data)

# Initialize the score
score = 0

# Loop through each question
for i, question in enumerate(quiz_data):
    print(f"\nQuestion {i+1}: {question['question']}")
    for j, answer in enumerate(question['answers']):
        print(f"{j+1}. {answer}")

    # Get the user's answer
    user_answer = input("Enter the number of your answer: ")

    # Check if the user's answer is correct
    if int(user_answer) - 1 == question['correct']:
        print("Correct!")
        score += 3
    else:
        print(f"Sorry, the correct answer is {question['answers'][question['correct']]}")
        score -= 1

# Print the final score
print(f"\nYour final score is: {score}")
```
